```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0    # mean
sigma = 1  # standard deviation

# Discretize the normal distribution at 16 points
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 16)
y = norm.pdf(x, mu, sigma)

# Plot the discretized normal distribution
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'bo-', label='Discretized Normal Distribution')
plt.fill_between(x, y, alpha=0.2)
plt.title('Normal Distribution Discretized at 16 Points')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
